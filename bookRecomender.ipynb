```python
popular_ratings_making = ratings.merge(books,on='ISBN')
```

```python
popular_ratings_making.head(3)
```

```python
popular_Number_Of_ratings_df = popular_ratings_making.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
popular_Number_Of_ratings_df.head(2)
```

```python
popular_Number_Of_ratings_df.rename(columns={'num_ratings':'num_of_ratings'},inplace=True)
popular_Number_Of_ratings_df.head(3)
```

```python
avg_rating_df = popular_ratings_making.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
avg_rating_df.head()
```

```python
popular_Book_df = popular_Number_Of_ratings_df.merge(avg_rating_df,on='Book-Title')
```

```python
popular_Book_df.head()
```

```python
popular_Book_df = popular_Book_df[popular_Book_df['num_of_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_Book_df.shape
```

```python
popular_Book_df = popular_Book_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_of_ratings','avg_rating']]
```

```python
popular_Book_df.head()
```

```python
popular_ratings_making.head(2)
```

```python
x = popular_ratings_making.groupby('User-ID').count()['Book-Rating'] > 200
best_users = x[x].index
```

```python
best_users
```

```python
filtered_rating = popular_ratings_making[popular_ratings_making['User-ID'].isin(best_users)]
filtered_rating.head()
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.head()
```

```python
mainData = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python

mainData.fillna(0,inplace=True)
```

```python
mainData.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(mainData)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):

    index = np.where(mainData.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == mainData.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
recommend('1984')
```
